```python
import libopencor as oc
```

```python
file = oc.File('../models/lorenz.omex')

if len(file.issues) != 0:
    print(file.issues[0].description)
else:
    print('File: all good!')
```

```python
document = oc.SedDocument(file)

if len(document.issues) != 0:
    print(document.issues[0].description)
else:
    print('Document: all good!')
```

```python
instance = document.instantiate()

if len(instance.issues) != 0:
    print(instance.issues[0].description)
else:
    print('Instance: all good!')
```

```python
instance.run()

if len(instance.issues) != 0:
    print(instance.issues[0].description)
else:
    print('Instance running: all good!')
```

```python
instance_task = instance.tasks[0]

x = instance_task.state(0)
y = instance_task.state(1)
z = instance_task.state(2)

%matplotlib inline

from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt

plt.axes(projection='3d').plot3D(x, y, z)
```
